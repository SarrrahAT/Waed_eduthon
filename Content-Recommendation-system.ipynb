
Collaborative Filtering Based  Recommeder System 


```python
x = book_rat.groupby('User-ID').count()['Book-Rating'] > 200
pl = x[x].index
```

```python
pl
```

```python
p = paper_rat.groupby('User-ID').count()['Paper-Rating'] > 200
plpaper = p[p].index
```

```python
plpaper
```

```python
fil_rat = book_rat[book_rat['User-ID'].isin(pl)]
fil_rat
```

```python
fil_paper_rat = paper_rat[paper_rat['User-ID'].isin(plpaper)]
fil_paper_rat
```

```python
y = fil_rat.groupby('Book-Title').count()['Book-Rating'] > 50
famous_book = y[y].index
```

```python
m = fil_paper_rat.groupby('title').count()['Paper-Rating'] > 15
famous_paper = m[m].index
```

```python
final_rat = fil_rat[fil_rat['Book-Title'].isin(famous_book)]
pt = final_rat.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
final_paper_rat = fil_paper_rat[fil_paper_rat['title'].isin(famous_paper)]
ptpaper = final_paper_rat.pivot_table(index='title',columns='User-ID',values='Paper-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
ptpaper.fillna(0,inplace=True)
```

```python
sim_score = cosine_similarity(pt)
sim_score
```

```python
sim_paper_score = cosine_similarity(ptpaper)
sim_paper_score
```

```python
# Assuming pt, ptpaper, sim_score, and sim_paper_score are already defined
# pt: DataFrame for books interactions
# ptpaper: DataFrame for papers interactions
# sim_score: similarity matrix for books
# sim_paper_score: similarity matrix for papers

def recommend_content_for_student(student_id, top_n=5):
    # Ensure that `pt`, `ptpaper`, `sim_score`, and `sim_paper_score` are correctly defined
    global pt, ptpaper, sim_score, sim_paper_score

    # Dictionary to store cumulative similarity scores
    content_scores = {}

    # Get the books the student has interacted with
    if student_id in pt.columns:
        student_book_interactions = pt.loc[:, student_id]
        interacted_books = student_book_interactions[student_book_interactions > 0].index.tolist()

        # Calculate similarity scores for each book the student has interacted with
        for book in interacted_books:
            book_index = np.where(pt.index == book)[0][0]
            for idx, score in enumerate(sim_score[book_index]):
                if pt.index[idx] not in interacted_books:  # Don't recommend already interacted books
                    if pt.index[idx] in content_scores:
                        content_scores[pt.index[idx]] += score
                    else:
                        content_scores[pt.index[idx]] = score

    # Get the papers the student has interacted with
    if student_id in ptpaper.columns:
        student_paper_interactions = ptpaper.loc[:, student_id]
        interacted_papers = student_paper_interactions[student_paper_interactions > 0].index.tolist()

        # Calculate similarity scores for each paper the student has interacted with
        for paper in interacted_papers:
            paper_index = np.where(ptpaper.index == paper)[0][0]
            for idx, score in enumerate(sim_paper_score[paper_index]):
                if ptpaper.index[idx] not in interacted_papers:  # Don't recommend already interacted papers
                    if ptpaper.index[idx] in content_scores:
                        content_scores[ptpaper.index[idx]] += score
                    else:
                        content_scores[ptpaper.index[idx]] = score

    # Sort content by their cumulative similarity score and select the top_n content
    recommended_content = sorted(content_scores.items(), key=lambda x: x[1], reverse=True)[:top_n]

    # Print the recommended content with comments
    print("Recommended Content:")
    for content, score in recommended_content:
        content_type = "Book" if content in pt.index else "Paper"
        print(f"{content_type}: {content} (Score: {score})")
```

```python
# Example usage with an actual student_id from your dataset
recommend_content_for_student(11676)  # Replace with actual student ID from the dataset
```
